#### Create train and test sets, create features for each of them

```python
# Creating dummies
df = df_test
dummies = pd.get_dummies(df[['inspection_status', 'inspection_category','boiler', 'refrigeration', 'elevator']])
newdf = pd.merge(df, dummies, left_index = True, right_index = True) 
# Create a building features dataframe
b_features = pd.DataFrame()
buildings = newdf.groupby(['property_group'])
b_features['num_failed'] = buildings.inspection_status_FAILED.sum()
b_features['num_passed'] = buildings.inspection_status_PASSED.sum()
b_features['total_inspected'] = b_features.num_failed + b_features.num_passed
b_features['num_complaint'] = buildings.inspection_category_COMPLAINT.sum()
b_features['response'] = b_features.num_failed > 0
b_features['boiler'] = buildings.boiler_y.sum() > 0
b_features['refrigeration'] = buildings.refrigeration_y.sum() > 0
b_features['elevator'] = buildings.elevator_y.sum() > 0
b_features_test = b_features
```

```python
df = df_train
dummies = pd.get_dummies(df[['inspection_status', 'inspection_category','boiler', 'refrigeration', 'elevator']])
newdf = pd.merge(df, dummies, left_index = True, right_index = True) 
# Create a building features dataframe
b_features = pd.DataFrame()
buildings = newdf.groupby(['property_group'])
b_features['num_failed'] = buildings.inspection_status_FAILED.sum()
b_features['num_passed'] = buildings.inspection_status_PASSED.sum()
b_features['total_inspected'] = b_features.num_failed + b_features.num_passed
b_features['num_complaint'] = buildings.inspection_category_COMPLAINT.sum()
b_features['response'] = b_features.num_failed > 0
b_features['boiler'] = buildings.boiler_y.sum() > 0
b_features['refrigeration'] = buildings.refrigeration_y.sum() > 0
b_features['elevator'] = buildings.elevator_y.sum() > 0
b_features_train = b_features
```

####Counting unique occurences within groupby: nunique()

```python

grouped = df[['violation_date', 'inspection_category','id']].groupby(['violation_date', 'inspection_category']).count()
```

```python
grouped.unstack(level = -1).plot(figsize = [15, 5])
```

```python
df['property_group'].unique().size
grouped_bd = df[['violation_date','inspection_category','property_group']].groupby(['violation_date','inspection_category'])['property_group'].nunique()
grouped_bd # unique buildings
```

```python
grouped_bd.unstack(level = -1).plot(figsize = [15, 5], title = 'daily violating building count ~ time')
```

```python
grouped_bd_m = grouped_bd.unstack(level = -1).resample('M', how='sum')
grouped_bd_m.plot(figsize = [15, 5], title = 'monthly violating building count ~ time')
```
